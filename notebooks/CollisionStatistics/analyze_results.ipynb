```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn')
```

```python
def parse_mbmr_file(filename, n_samples = 1002, n_mbmr = 5):
    vals = []
    with open(filename) as f:
        for line in f:
            if not line[0]==';' and not len(line) < 3:
                vals.append(float(line))
            
    vals = np.array(vals)
    
    vals_segmented = [vals[i*n_samples:(i+1)*n_samples-1] for i in range(n_mbmr)]
    return(vals_segmented)

def plot_mbmr(mbmr_dat):
    n_mbmr = np.shape(mbmr_dat)[0]
    plt.figure(figsize=(10,6))
    for i in range(n_mbmr):
        plt.plot(np.linspace(0,1,len(mbmr_dat[i])), mbmr_dat[i]*1,label = 'mbmr_'+str(i+1))
        
    plt.legend()
    plt.xlabel("probability")
    plt.ylabel("distance")

plot_mbmr(parse_mbmr_file('cs_icdf_2020_02_27_001_test.dat'))
```
